```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import square  # Función para generar señales cuadradas
from scipy.fft import fft, fftfreq  
```

```python
# Parámetros de la señal
f = 5  # Frecuencia de la señal en Hertz (Hz)
amplitud = 1  # Amplitud máxima de la señal
ciclos = 2  # Número de ciclos completos de la señal cuadrada
```

```python
# CASO N1: 100 muestras
N1 = 100  # Número total de muestras para el primer caso
fs1 = f * N1 / ciclos  # Frecuencia de muestreo, definida por la relación entre frecuencia, muestras y ciclos
T1 = 1/fs1  # Período de muestreo, inversa de la frecuencia de muestreo

# Creación del vector temporal con N1 muestras, desde 0 hasta duración total sin incluir punto final
t1 = np.linspace(0, ciclos/f, N1, endpoint=False)
# Generación de la señal cuadrada normalizada entre 0 y 1
senal1 = 0.5 * (square(2 * np.pi * f * t1, duty=0.5) + 1)
```

```python
# Gráfica temporal para N1 muestras

# Configuración de la gráfica
plt.figure(figsize=(14, 12))  # Definición del tamaño total de la figura

plt.subplot(4, 1, 1)  # Definición del primer gráfico en la figura (arriba)
plt.plot(t1, senal1)  # Gráfica de la señal cuadrada en el tiempo
plt.title("Señal en tiempo con N1 = 100 muestras")  # Título del gráfico
plt.xlabel("Tiempo (s)")  # Etiqueta del eje horizontal
plt.ylabel("Amplitud")  # Etiqueta del eje vertical

# Ajuste de gráficos para una mejor visualización
plt.tight_layout()
# Mostrar todas las gráficas
plt.show()
```

```python
# Cálculo de la FFT para la señal generada (N1 muestras)
yf1 = fft(senal1)  # Aplicación de la transformada rápida de Fourier
xf1 = fftfreq(N1, T1)[:N1//2]  # Cálculo de las frecuencias para representar el espectro
```

```python
# Espectro de frecuencias (FFT) para N1 muestras

# Configuración de la gráfica
plt.figure(figsize=(14, 12))  # Definición del tamaño total de la figura

plt.subplot(4, 1, 2)  # Segundo gráfico
plt.plot(xf1, np.abs(yf1[:N1//2])) # Magnitud del espectro FFT
plt.title("Espectro FFT con N1 = 100 muestras")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("|FFT|")

# Ajuste de gráficos para una mejor visualización
plt.tight_layout()
# Mostrar todas las gráficas
plt.show()
```

```python
# CASO N2: 200 muestras
N2 = 200  # Número total de muestras para el segundo caso
fs2 = f * N2 / ciclos  # Frecuencia de muestreo para el segundo caso
T2 = 1/fs2  # Período de muestreo correspondiente al segundo caso

# Creación del vector temporal con N2 muestras
t2 = np.linspace(0, ciclos/f, N2, endpoint=False)
# Generación de la señal cuadrada normalizada entre 0 y 1 para N2 muestras
senal2 = 0.5 * (square(2 * np.pi * f * t2, duty=0.5) + 1)
```

```python
# Gráfica temporal para N2 muestras

# Configuración de la gráfica
plt.figure(figsize=(14, 12))  # Definición del tamaño total de la figura

plt.subplot(4, 1, 3)  # Tercer gráfico
plt.plot(t2, senal2)
plt.title("Señal en tiempo con N2 = 200 muestras")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")

# Ajuste de gráficos para una mejor visualización
plt.tight_layout()
# Mostrar todas las gráficas
plt.show()
```

```python
# Cálculo de la FFT para la señal generada (N2 muestras)
yf2 = fft(senal2)  # Transformada rápida de Fourier para el segundo caso
xf2 = fftfreq(N2, T2)[:N2//2]  # Frecuencias correspondientes al espectro del segundo caso
```

```python
# Espectro de frecuencias (FFT) para N2 muestras

# Configuración de la gráfica
plt.figure(figsize=(14, 12))  # Definición del tamaño total de la figura

plt.subplot(4, 1, 4)  # Cuarto gráfico
plt.plot(xf2, np.abs(yf2[:N2//2]))
plt.title("Espectro FFT con N2 = 200 muestras")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("|FFT|")

# Ajuste de gráficos para una mejor visualización
plt.tight_layout()
# Mostrar todas las gráficas
plt.show()
```

```python
# Gráfica temporal superpuesta

# Configuración de la gráfica
plt.figure(figsize=(14, 12))  # Definición del tamaño total de la figura

plt.subplot(6, 1, 5)
plt.plot(t1, senal1, label="N1 = 100 muestras", color='blue')
plt.plot(t2, senal2, label="N2 = 200 muestras", color='orange')
plt.title("Comparación temporal de señales")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.legend()

# Ajuste de gráficos para una mejor visualización
plt.tight_layout()
# Mostrar todas las gráficas
plt.show()
```

```python
# Gráfica FFT superpuesta

# Configuración de la gráfica
plt.figure(figsize=(14, 12))  # Definición del tamaño total de la figura

plt.subplot(6, 1, 6)
plt.plot(xf1, np.abs(yf1[:N1//2]), label="N1 = 100 muestras", color='blue')
plt.plot(xf2, np.abs(yf2[:N2//2]), label="N2 = 200 muestras", color='orange')
plt.title("Comparación del espectro FFT")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("|FFT|")
plt.legend()

# Ajuste de gráficos para una mejor visualización
plt.tight_layout()
# Mostrar todas las gráficas
plt.show()
```

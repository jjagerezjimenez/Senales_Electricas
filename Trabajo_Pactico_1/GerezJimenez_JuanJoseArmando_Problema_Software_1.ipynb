APARTADO B

Se que para calcular los valores de la senal debo usar las definiciones vistas en clases:




Como estamos trabajando con senales rectangulares, podemos simplificar los calculos de las integrales usando calculos basicos de areas.

```python
# Se√±al A:
# Defino valores para el calculo de las areas
 
V_alto_A, t_alto_A = 1.9, 0.2
V_bajo_A, t_bajo_A = 0.9, 1.8
T = 2
```

```python
Vmedio_A = (V_alto_A*t_alto_A + V_bajo_A*t_bajo_A) / T
```

```python
Vcuad_A = (V_alto_A**2*t_alto_A + V_bajo_A**2*t_bajo_A) / T
```

```python
VRMS_A = np.sqrt(Vcuad_A)
```

```python
VAC_A = np.sqrt(Vcuad_A**2 - Vmedio_A**2)
```

```python
print(f'Se√±al A: Vmedio={Vmedio_A:.2f}, VRMS={VRMS_A:.3f}, VAC={VAC_A:.2f}')
```

```python
# Se√±al B:
V_alto_B, t_alto_B = 1.1, 1.8
V_bajo_B, t_bajo_B = 0.1, 0.2
```

```python
Vmedio_B = (V_alto_B*t_alto_B + V_bajo_B*t_bajo_B) / T
```

```python
Vcuad_B = (V_alto_B**2*t_alto_B + V_bajo_B**2*t_bajo_B) / T
```

```python
VRMS_B = np.sqrt(Vcuad_B)
```

```python
VAC_B = np.sqrt(Vcuad_B**2 - Vmedio_B**2)
```

```python
print(f'Se√±al B: Vmedio={Vmedio_B:.2f}, VRMS={VRMS_B:.3f}, VAC={VAC_B:.2f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

#  PARMETROS BSICOS 
Ts = 0.5e-3         # Perodo de muestreo [s]  (0,5 ms  Fs = 2 kHz)
Fs = 1 / Ts         # Frecuencia de muestreo [Hz]
f0 = 3e3            # Frecuencia de la seal analgica [Hz]

# Queremos mostrar exactamente 8 ciclos de la seal analgica
ciclos_a_mostrar = 8
t_fin = ciclos_a_mostrar / f0      # Duracin total para esos 8 ciclos

#  VECTORES DE TIEMPO 
t_muestras = np.arange(0, t_fin, Ts)         # Instantes de muestreo
t_continuo = np.linspace(0, t_fin, 1000)     # Tiempo fino para la seal continua

#  SEALES (muestras y forma continua) 
seno_muestras = np.sin(2 * np.pi * f0 * t_muestras)
coseno_muestras = np.cos(2 * np.pi * f0 * t_muestras)

seno_continuo = np.sin(2 * np.pi * f0 * t_continuo)
coseno_continuo = np.cos(2 * np.pi * f0 * t_continuo)

#  GRFICA DE LA SENOIDE 
plt.figure()
plt.plot(t_continuo * 1e3,
         seno_continuo,
         color='r',
         label='Seal analgica (seno)',
         linestyle='-')

plt.stem(t_muestras * 1e3,
         seno_muestras,
         basefmt=" ",
         label='Muestras')

plt.title('Muestreo de la senoide de 3 kHz (8 ciclos)')
plt.xlabel('Tiempo [ms]')
plt.ylabel('Amplitud')
plt.grid(True)
plt.legend()

#  GRFICA DEL COSENO 
plt.figure()
plt.plot(t_continuo * 1e3,
         coseno_continuo,
         label='Seal analgica (coseno)',
         linestyle='-')

plt.stem(t_muestras * 1e3,
         coseno_muestras,
         linefmt='r',
         markerfmt='ro',
         basefmt=" ",
         label='Muestras')

plt.title('Muestreo del coseno de 3 kHz (8 ciclos)')
plt.xlabel('Tiempo [ms]')
plt.ylabel('Amplitud')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
